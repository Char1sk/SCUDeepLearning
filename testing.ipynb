```python
import numpy as np
from scipy.io import loadmat
mat = loadmat("./final.mat")
trainOut = mat['trainOut']
testOut = mat['testOut']
trainLabel = mat['trainLabel']
testLabel = mat['testLabel']
print(trainOut[:,0:6])
print(trainLabel[:,0:6])
print((np.argmax(trainOut[:,0:6], axis=0)+1)%10)
# print(testOut)
```

```python
from sklearn import metrics

print("train dataset classification report")
print(metrics.classification_report(np.argmax(trainOut, axis=0), np.argmax(trainLabel, axis=0)))

print('-'*60)
print('-'*60)

print("test dataset classification report")
print(metrics.classification_report(np.argmax(testOut, axis=0), np.argmax(testLabel, axis=0)))
```

```python
from sklearn import metrics
from plot_confusion_matrix import plot_confusion_matrix

cm = metrics.confusion_matrix(np.argmax(trainLabel, axis=0), np.argmax(trainOut, axis=0))
plot_confusion_matrix(cm, range(10), title='train dataset Confusion matrix')

cm = metrics.confusion_matrix(np.argmax(testLabel,  axis=0), np.argmax(testOut,  axis=0))
plot_confusion_matrix(cm, range(10), title='test  dataset Confusion matrix')
```
